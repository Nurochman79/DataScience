```python
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
df_sy = pd.read_excel('/content/data.xlsx', names=["Item"], header=None)
df_sy.head(17)
```

```python
df_sy.shape
```

```python
records=[]
for i in range (df_sy.shape[0]):
    records.append([str(df_sy.values[i,j]).split(',') for j in range(df_sy.shape[1])])

trx = [[] for trx in range (len(records))]
for i in range (len(records)):
    for j in records[i][0]:
        trx[i].append(j)
trx
```

```python
from mlxtend.preprocessing import TransactionEncoder
trans= TransactionEncoder()
df_trans= trans.fit(trx).transform(trx)
df=pd.DataFrame(df_trans, columns=trans.columns_)
df.head(14)
```

```python
df.shape
```

```python
frequent_itemsets=apriori(df, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets.count()['itemsets']
```

```python
frequent_itemsets.nlargest(n=10, columns="support")
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(frequent_itemsets,metric='lift', min_threshold=1)
rules.sort_values(by=['support'], ascending=False)
```
