```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/heart1.csv')
data.head()
```

```python
data.value_counts()
```

```python
avrg_age=data.groupby('target').mean()
print(avrg_age)
```

```python
target = data['target']
```

```python
target1=data['target'].value_counts()
target1.shape
```

```python
target1.value_counts()
```

```python
plt.pie(target1, labels=target1, autopct='%d%%',startangle=200)
plt.axis('equal')
plt.title('Persentase Kategori Target')
plt.show()
```

```python
sum_age=data['age'].value_counts()
sum_age.head(50)
```
